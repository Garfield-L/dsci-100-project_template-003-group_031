Part 3. Predictive Question Choice

Predictive Question: "Can players information (ie age, gender, experience level) predict which players will have higher session frequency than average?"

My Methods and Plans:
Firstly, I would use knn classification, it is appropriate because we are predicting a two-sides outcome (higher or lower than the average), by using categorical and numerical predictors, therefore this can be a good example of knn model. I will test different values of K (neighbors) and choose the one that is most accurate. I will separate my data into 70% for training and 30% for testing, and use the training set to build the model and the testing set to check how it works in real life. This method allows me to determine if a player characteristics can reliably identify as a high frequency people or not.

To begin, we will write code to create our two sides target variable (high frequency player) and prepare the predictor variables (age, gender, experience) for KNN classification with some basic data wrangling.

Part 4. Wrangling and Exploratory Analysis

```r
# Part 4.1: Create two sides Response Variable for High Frequency Players

# Calculate total sessions per player
session_counts <- sessions |>
  group_by(hashedEmail) |>
  summarize(total_sessions = n())

# Calculate the median sessions
median_sessions <- session_counts |>
  summarize(median_sessions = median(total_sessions)) |>
  pull(median_sessions)

print(median_sessions)

# Create the high frequency players
session_counts <- session_counts |>
  mutate(high_frequency_player = total_sessions > median_sessions)

# Check the distribution
frequency_summary <- session_counts |>
  group_by(high_frequency_player) |>
  summarize(count = n())

print("Distribution of high vs low frequency players:")
print(frequency_summary)
```

Above thing shows that the median session count is only 1, with 44 players classified as high frequency (above median) and 81 as low frequency.

```r
# Get the list of high frequency player emails
high_freq_emails <- session_counts |>
  filter(high_frequency_player == TRUE) |>
  select(hashedEmail)

# Create analysis dataset and add frequency
analysis_data <- players |>
  select(hashedEmail, Age, gender, experience, played_hours)|>
  mutate(high_frequency_player = hashedEmail %in% high_freq_emails$hashedEmail)

# Results
print("High frequency player count in final data:")
analysis_data |>
  group_by(high_frequency_player) |>
  summarize(count = n())

head(analysis_data,7)
```

Above things show that it contains 196 players with 44 classified as high frequency people and 152 as low frequency.

```r
# Part 4.3: Exploratory Analysis

# Check experience levels
analysis_data |>
  group_by(experience, high_frequency_player) |>
  summarize(count = n())

# Check average age by frequency
print("Average age by frequency category:")
analysis_data |>
  group_by(high_frequency_player) |>
  summarize(avg_age = mean(Age))
```

Above things show that high frequency players tend to be younger (average age 19.6) and are distributed across all experience levels (since number of players are distributed in every experience), Amateur players has the highest counts in both frequency categories.

```r
# Part 4.4: Prepare for KNN

# Select only the variables we need for modeling
model_data <- analysis_data |>
  select(Age, gender, experience, high_frequency_player)

print("Final model dataset ready for KNN:")
print(model_data)
print("Target variable distribution:")
table(model_data$high_frequency_player)
```

The final dataset contains 196 observations with 44 high frequency players and 152 low frequency players. Age, gender, and experience are as predictors for KNN classification in the next part.

Part 5. KNN Classification and Visualizations

```r
library(tidymodels)
library(ggplot2)

# Create recipe
game_recipe <- recipe(high_frequency_player ~ Age + gender + experience, 
                      data = model_data) |>
  step_scale(all_numeric_predictors()) |>
  step_center(all_numeric_predictors())

# KNN model
knn_spec <- nearest_neighbor(weight_func = "rectangular", 
                             neighbors = tune()) |>
  set_engine("kknn") |>
  set_mode("classification")

# Create workflow
knn_workflow <- workflow() |>
  add_recipe(game_recipe) |>
  add_model(knn_spec)


print("KNN workflow created")
```

```r
# Visualization Analysis

# Plot 1: Age vs High Frequency Players
plot_age <- ggplot(model_data, aes(x = Age, fill = high_frequency_player)) +
  geom_histogram(position = "dodge", bins = 15) +
  labs(title = "Age Distribution by Player Frequency",
       x = "Age", 
       y = "Number of Players") +
  theme_minimal()

# Plot 2: Gender vs High Frequency Players
plot_gender <- model_data |>
  ggplot(aes(x = gender, fill = high_frequency_player)) +
  geom_bar(position = "dodge") +
  labs(title = "Gender Distribution by Player Frequency",
       x = "Gender", 
       y = "Number of Players") +
  theme_minimal()

# Plot 3: Experience Level vs High Frequency Players 
plot_experience <- model_data |>
  ggplot(aes(x = experience, fill = high_frequency_player)) +
  geom_bar(position = "dodge") +
  labs(title = "Experience Levels by Player Frequency",
       x = "Experience Level", 
       y = "Number of Players") +
  theme_minimal()

# Plot 4: Played Hours vs High Frequency Players
plot_hours <- analysis_data |>
  filter(played_hours < 50) |>
  ggplot(aes(x = played_hours, fill = high_frequency_player)) +
  geom_histogram(position = "dodge", bins = 20) +
  labs(title = "Played Hours Distribution by Player Frequency",
       x = "Hours Played", 
       y = "Number of Players") +
  theme_minimal()

print(plot_age)
print(plot_gender)
print(plot_experience)
print(plot_hours)
```

```r
# Regression System for Age and Played Hours

# Set seed
set.seed(100)

# Prepare data
regression_data <- analysis_data |>
  select(Age, played_hours)

# Create simple recipe without variables
session_recipe <- recipe(played_hours ~ Age, data = regression_data) |>
  step_scale(all_numeric_predictors()) |>
  step_center(all_numeric_predictors())

# Split data for regression
session_split <- initial_split(regression_data, prop = 0.75, strata = played_hours)
session_training <- training(session_split)
session_testing <- testing(session_split)

print(paste("Training rows:", nrow(session_training)))
print(paste("Testing rows:", nrow(session_testing)))

best_k <- 5  # Using a reasonable value

print(paste("Using fixed neighbors:", best_k))

# Fit model with fixed K
best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = best_k) |>
  set_engine("kknn") |>
  set_mode("regression")

best_fit <- workflow() |>
  add_recipe(session_recipe) |>
  add_model(best_spec) |>
  fit(data = session_training)

# Check modle
session_predictions <- best_fit |>
  predict(session_testing) |>
  bind_cols(session_testing)

session_metrics <- session_predictions |>
  metrics(truth = played_hours, estimate = .pred)

print(session_metrics)

# Create visualization
session_plot <- ggplot(session_predictions, aes(x = Age, y = played_hours)) +
  geom_point(alpha = 0.6) +  
  geom_line(aes(y = .pred), color = "red", size = 1) + 
  labs(x = "Age",
       y = "Hours Played",
       title = "KNN Regression: Predicting Play Hours from Player Age")

print(session_plot)
```

This regression results above show a weak predictive power with a high RMSE (33.9) and a near-zero R-squared (0.002). This means there is no clear predictive relationship between factors like Age and Play hours. I only chose k=5 instead of tuning because the errors (like some NAs) made cross-validation not very reliable, so I choose a fixed number, and 5 is a moderate k value with stability (also because I didn't learn tune thing when I finished this). The prediction on the plot is straight line (a bit underfitting) and this is because KNN regression is the average of points, and since most players actually play a little amount of time no matter their age. Some other factors, like Experience Level or Gender, already showed a clear relationship in my visualizations before, which are already the conclusions of my questions. Therefore, I believe this regression can't effectively lead to the answer.

```r
# Linear Regression for Experience Level and Play Hours

# Set seed
set.seed(100)

# Prepare data
exp_regression_data <- analysis_data |>
  select(experience, played_hours) |>
  filter(!is.na(experience))

# Create recipe
exp_recipe <- recipe(played_hours ~ experience, data = exp_regression_data)

# Split data
exp_split <- initial_split(exp_regression_data, prop = 0.75, strata = played_hours)
exp_training <- training(exp_split)
exp_testing <- testing(exp_split)

# Use linear regression - appropriate for categorical predictors
linear_spec <- linear_reg() |>
  set_engine("lm") |>
  set_mode("regression")

# Fit linear model
linear_fit <- workflow() |>
  add_recipe(exp_recipe) |>
  add_model(linear_spec) |>
  fit(data = exp_training)

# Evaluate model
exp_predictions <- linear_fit |>
  predict(exp_testing) |>
  bind_cols(exp_testing)

exp_metrics <- exp_predictions |>
  metrics(truth = played_hours, estimate = .pred)

print("Linear Regression Performance (Experience vs Play Hours):")
print(exp_metrics)

exp_summary <- exp_regression_data |>
  group_by(experience) |>
  summarize(avg_played_hours = mean(played_hours))

exp_plot <- ggplot(exp_summary, aes(x = experience, y = avg_played_hours)) +
  geom_col(fill = "lightgreen", alpha = 0.7) +
  labs(x = "Experience Level",
       y = "Average Hours Played",
       title = "Average Play Hours by Experience Level") +
  theme_minimal()

print(exp_plot)
```

This time I choose linear regression for experience levels and hours played because it effectively helps me to see these categorical predictors. It gives clear results showing how each category affects play hours. The results show moderate outcomes with r square of 0.109, telling that experience level explains about 10% of the variation in play hours. The visualization in the end shows that Regular players dominate in average play hours, then they are Amateur and Pros. This partly answer my broad question by pointing out that Regular players as the main kind of players who contribute larger amounts of data.
